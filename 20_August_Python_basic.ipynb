```python
"""  42. Encode a string using a Caesar cipher.
xplanation = formula for encryption = (index(char)+shift_value)%26"""
alphabet=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
def encryption(plain_text, shift_key):
    cipher_text=""
    for i in plain_text:
        position=alphabet.index(i)
        new_position=(position+shift_key)%26
        cipher_text+=alphabet[new_position]
    print("the cipher text for plain text is ",cipher_text)
encryption("jenny",3)
```

```python
"""  43. Decode a Caesar cipher encoded string.
explanation = formula for decryption = (index(char)-shift_value)%26 """
alphabet=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
def decryption(cipher_text, shift_key):
    plain_text=""
    for i in cipher_text:
        position=alphabet.index(i)
        new_position=(position-shift_key)%26
        plain_text+=alphabet[new_position]
    print("the plain text for the respected cipher text is ",plain_text)
decryption("mhqqb",3)
    
```

```python
"""  44. Find the most frequent word in a string.
explantion= split the string into words , then counter() method will count the frequency of each word
then most_common() method will return tuple each tuple has word with its frequency, We extract the word from the most common tuple.
and then return the most_common_word.
"""
import re
from collections import Counter
input_string=input("enter the string")


def most_frequent_word(input_string):
    words = input_string.split()
    word_counter = Counter(words)
    most_common_word = word_counter.most_common(1)[0][0]
    return most_common_word


print("most frequent word is",most_frequent_word(input_string))
```

```python
"""  50. Convert a CSV string to a list of lists."""
def csv_string_to_list(csv_string):
    # Split the CSV string into lines
    lines = csv_string.split('\n')
    print(lines)

    # Initialize an empty list to store the result
    result = []

    for i in lines:
        # Split each line into fields using comma as the delimiter
        fields = i.split(',')

        # Append the fields to the result list as a row
        result.append(fields)

    return result



# Convert CSV string to a list of lists
list_of_lists = csv_string_to_list("Name,Age, Country\nJohn, 25, USA\nAlice, 30, Canada\nBob, 22, UK")

# Print the result
for row in list_of_lists:
    print(row)
```

```python
"""21. Create a set of the first 5 prime numbers."""
s=set()
i=2
while(len(s)<=5-1):
    a=0
    for j in range(2,i):
        if(i%j==0):  #if i and j is divisible then it is not prime number , make a=1 and break the loop
            a=1
            break
    if(a==0):  #if a is 0 that means i and j is not divisible -> prime number , add to set 
        s.add(i)
    i+=1
print("the set of 5 prime numbers ",s)
```
